```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 加载txt文件到内存
f = open("ex1data1.txt", "r")
# redline 方法每次读取一行，返回一行的读取结果，下一次调用f.readlin将从下一行开始
l = f.readline()
# 初始化生成一个二维矩阵数据
xy = np.zeros((2, 97))
print("xy shape:", xy.shape)
# xy shape: (2, 97)
# 定义一个变量 a = 0
a = 0
# 开始进行 while 循环, 如果l有值，就继续执行
while l:
    # 将l转换为数组['6.1101', '17.592\n']
    l2 = l.split(",")
    # f1就是数组的第一个元素，f2就是数组的第二个元素
    f1 = float(l2[0])
    f2 = float(l2[1])
    # x轴和y轴的数据 xy[0, 0] = '6.1101'  xy[1, 0] = '17.592\n'
    xy[0, a] = f1
    xy[1, a] = f2
    # 每次a加1
    a += 1
    # 读取下一行
    l = f.readline()
# 关闭读取文件的内存空间
f.close()

# 设置图标的背景颜色
plt.figure('Scatter', facecolor='#0f0')
# 设置title size
plt.title('Scatter', fontsize=20)
# x轴描述
plt.xlabel('Population of City in 10,000s', fontsize=7)  # 水平坐标
# y轴描述
plt.ylabel('Profit in $10,000', fontsize=7)  # 垂直坐标
# tick字体大小
plt.tick_params(labelsize=10)
# **linestyle ?*也可以用ls来代替linestyle， 设置网格线的风格，是连续实线，虚线或者其它不同的线条。 | ‘-’ | ‘–’ | ‘-.’ | ‘:’ | ‘None’ | ‘’ | ‘’]
plt.grid(linestyle=':')

# x，y：表示的是大小为(n,)的数组，也就是我们即将绘制散点图的数据点
# s
# cmap
# alpha
# marker
# 讲数据点描绘到图上
plt.scatter(xy[0, :], xy[1, :], s=60, cmap='jet_r', alpha=0.5, marker='o')
# scatter画散点图  c=d 颜色=距离要与cmap(颜色映射)同用,正向映射(jet)反向(jet_r),alpha透明度,marker图形（*，d..）
plt.show()

# 梯度下降函数
def gradient_descent(train_data):
    # 标识矩阵第一维的长度 m = 97
    m = train_data.shape[1]
    alpha = 0.01                    # learning rate
    theta0 = 0.
    theta1 = 0.
    J_theta = 0.
    for i in range(15000):
        theta0_tmp = theta0
        theta0 = theta0 - alpha * 1 / m * \
            np.sum((theta0 + theta1 * train_data[0, :] - train_data[1, :]) * 1)
        theta1 = theta1 - alpha * 1 / m * \
            np.sum((theta0_tmp + theta1 *
                   train_data[0, :] - train_data[1, :]) * train_data[0, :])
        J_theta = 1 / (2 * m) * np.sum(np.square(theta0 +
                       theta1 * train_data[0, :] - train_data[1, :]))
        print(J_theta)

    # print(theta0, theta1)
    return theta0, theta1
    
# print('gradient_descent(xy)===>', xy)
theta = gradient_descent(xy)
# print("theta =", theta)
# print(type(theta))
```

```python
plt.figure('Scatter',facecolor='lightgray')
plt.title('Scatter',fontsize = 20)
plt.xlabel('Population of City in 10,000s',fontsize = 7) #水平坐标
plt.ylabel('Profit in $10,000',fontsize = 7)# 垂直坐标
plt.tick_params(labelsize = 10)
plt.grid(linestyle = ':')
plt.scatter(xy[0,:],xy[1,:],s = 60, cmap = 'jet_r',alpha = 0.5, marker = 'x')
#scatter画散点图  c=d 颜色=距离要与cmap(颜色映射)同用,正向映射(jet)反向(jet_r),alpha透明度,marker图形（*，d..）
x = np.linspace(np.min(xy[0, :]), np.max(xy[0, :]), 100)
y = theta[0] + theta[1] * x
plt.plot(x, y, c='black')
plt.show()
```
