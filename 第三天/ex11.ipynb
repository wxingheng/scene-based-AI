```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# txt
f = open("ex1data1.txt", "r")
# redline f.readlin
l = f.readline()
# 
xy = np.zeros((2, 97))
print("xy shape:", xy.shape)
# xy shape: (2, 97)
#  a = 0
a = 0
#  while , l
while l:
    # l['6.1101', '17.592\n']
    l2 = l.split(",")
    # f1f2
    f1 = float(l2[0])
    f2 = float(l2[1])
    # xy xy[0, 0] = '6.1101'  xy[1, 0] = '17.592\n'
    xy[0, a] = f1
    xy[1, a] = f2
    # a1
    a += 1
    # 
    l = f.readline()
# 
f.close()

# 
plt.figure('Scatter', facecolor='#0f0')
# title size
plt.title('Scatter', fontsize=20)
# x
plt.xlabel('Population of City in 10,000s', fontsize=7)  # 
# y
plt.ylabel('Profit in $10,000', fontsize=7)  # 
# tick
plt.tick_params(labelsize=10)
# **linestyle ?*lslinestyle  | - |  | -. | : | None |  | ]
plt.grid(linestyle=':')

# xy(n,)
# s
# cmap
# alpha
# marker
# 
plt.scatter(xy[0, :], xy[1, :], s=60, cmap='jet_r', alpha=0.5, marker='o')
# scatter  c=d =cmap(),(jet)(jet_r),alpha,marker*d..
plt.show()

# 
def gradient_descent(train_data):
    #  m = 97
    m = train_data.shape[1]
    alpha = 0.01                    # learning rate
    theta0 = 0.
    theta1 = 0.
    J_theta = 0.
    for i in range(15000):
        theta0_tmp = theta0
        theta0 = theta0 - alpha * 1 / m * \
            np.sum((theta0 + theta1 * train_data[0, :] - train_data[1, :]) * 1)
        theta1 = theta1 - alpha * 1 / m * \
            np.sum((theta0_tmp + theta1 *
                   train_data[0, :] - train_data[1, :]) * train_data[0, :])
        J_theta = 1 / (2 * m) * np.sum(np.square(theta0 +
                       theta1 * train_data[0, :] - train_data[1, :]))
        print(J_theta)

    # print(theta0, theta1)
    return theta0, theta1
    
# print('gradient_descent(xy)===>', xy)
theta = gradient_descent(xy)
# print("theta =", theta)
# print(type(theta))
```

```python
plt.figure('Scatter',facecolor='lightgray')
plt.title('Scatter',fontsize = 20)
plt.xlabel('Population of City in 10,000s',fontsize = 7) #
plt.ylabel('Profit in $10,000',fontsize = 7)# 
plt.tick_params(labelsize = 10)
plt.grid(linestyle = ':')
plt.scatter(xy[0,:],xy[1,:],s = 60, cmap = 'jet_r',alpha = 0.5, marker = 'x')
#scatter  c=d =cmap(),(jet)(jet_r),alpha,marker*d..
x = np.linspace(np.min(xy[0, :]), np.max(xy[0, :]), 100)
y = theta[0] + theta[1] * x
plt.plot(x, y, c='black')
plt.show()
```
